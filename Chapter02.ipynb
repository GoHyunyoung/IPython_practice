```python
def get_counts(sequence):
    counts={}
    for x in sequence:
        if x in counts:
            counts[x]+=1
        else:
            counts[x]=1
    return counts
```

```python
from collections import defaultdict
```

```python
counts=get_counts(time_zone)
```

```python
counts['America/New_York']
```

```python
def top_counts(count_dict,n=10):  #def로 함수정의 #가장많이 등장하느 상의 10개의 tz
    value_key_pairs=[(count,tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
#  list[0:3] = 인덱스 0부터 2까지
# list[:3] = 인덱스 0부터 2까지
# list[0:] = 인덱스 0부터 끝까지
# list[:-1] = list[:6] 
```

```python
top_counts(counts)
```

### 2.2 MovieLens의 영화평점 데이터

```python
import pandas as pd
```

```python
unames = ['user_id','gender','age','occupation','zip']
users=pd.read_table('https://raw.githubusercontent.com/pydata/pydata-book/master/ch02/movielens/users.dat',sep='::',header=None,names=unames)

rnames = ['user_id','movie_id','rating','timestamp']
ratings = pd.read_table('https://raw.githubusercontent.com/pydata/pydata-book/master/ch02/movielens/ratings.dat',sep='::',header=None,names=rnames)

mnames=['movie_id','title','genres']
movies = pd.read_table('https://raw.githubusercontent.com/pydata/pydata-book/master/ch02/movielens/movies.dat',sep='::',header=None,names=mnames)
```

```python
users[:5]
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
ratings
```

```python
#3가지의 데이터를 하나로 병합 -> pandas의 merge 함수를 이용
data=pd.merge(pd.merge(ratings,users),movies)
```

```python
data
```

```python
data.ix[0]
#행의 값을 확인할때는 ix사용
```

```python
mean_ratings=data.pivot_table('rating',index='title', columns='gender',aggfunc='mean')
#rows => index, cols => columns 로 변경(가로축 title, 세로축 gender, aggregatefunc deafult  = mean)
```

```python
mean_ratings[:5]
```

```python
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250] 
#250건 이상의 평점 정보가 있는 영화의 title만 가져오기
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings.ix[active_titles]
```

```python
#여성에게 높은 평점을 받은 영화목록 확인(F열을 내림차순으로 정렬)
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
```

### 2.2.1 평점 차이 구하기

```python
#남겨의 호불호가 갈리는 영화 찾기(mean_rating에 평균 평점의 차이를 담을수 있는 열을 하나 추가하여 그 열을 기준으로 정렬)
mean_ratings['diff']=mean_ratings['M']-mean_ratings['F']
```

```python
names.groupby('sex').births.sum()
#출생수를 모두합한 값을 해당연도에 전체 출생수로 가정
```

```python
years = range(1880,2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path='https://raw.githubusercontent.com/pydata/pydata-book/master/ch02/names/yob%d.txt' % year
    frame = pd.read_csv(path,names=columns)
    
    frame['year'] = year
    pieces.append(frame)

#1880년부터 2010년까지의 데이터를 하나의 DataFrame으로 종합
names = pd.concat(pieces,ignore_index=True)
```

```python
names
```

```python
total_births = names.pivot_table('births',index='year',columns='sex',aggfunc=sum)
#연도나 성별에 따른 데이터 정리
```

```python
total_births.tail()
#tail()은 마지막 5개 데이터만 추출
```

```python
total_births.plot(title='Total births by sex and year')
#이름이 각 출생수에서 차지하는 비율계산해보기
```

```python
def add_prop(group):
    births=group.births.astype(float)
    
    group['prop']=births/ births.sum()
    return group

#group에 prop 열을 추가하고 연산결과를 입력
```

```python
names = names.groupby(['year','sex']).apply(add_prop);
#names에 'prop' 추가
```

```python
names
```

```python
#prop열의 합이 1이 맞느지 확인하는 sanity check
np.allclose(names.groupby(['year','sex']).prop.sum(),1)
##### 'groupby' #####
```

```python
#각 연도별 성별에 따른 빈도수가 가장 높은 이름 1,000개 추출
def get_top1000(group):
    return group.sort_index(by='births',ascending=False)[:1000]
```

```python
grouped = names.groupby(['year','sex'])
```

```python
top1000 = grouped.apply(get_top1000)
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop',ascending = False)
    return group.prop.cumsum().searchsorted(q)[0] +1
```

```python
diversity = top1000.groupby(['year','sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
```
